```python
import json
import pandas as pd
import numpy as np
import itertools
```

```python
# modify the Database name, components, phases, element, comments based on your system
df = pd.read_excel("Hf-Mo_Taylor_1961.xlsx")
components= ['HF', 'MO', 'VA']
phases = ["BCC_A2", "HCP_A3", "LAVES_C15", "LAVES_C36", "LIQUID"]
element = "HF" # concentration element in the phase diagram
comment = "null"
reference = df.REF_JSON[0]


#create a Temperature list  
temperature = []
for i in df.Temperature:
    i= float(round(i, 3)) # Temperature round up to 3 decimal
    temperature.append(i)
#create a phase boundary data list 
values = []
for (a, b, c) in zip(df[element], df["Phase_1"], df["Phase_2"]):
    a= float(round(a, 3)) # Concentration round up to 3 decimal
    values.append([[b, [element], [a]], [c, [element], [None]]])
#print(values)
dictionary1 ={
  "components": components,
  "phases": phases,
  "conditions": {
      "P": 101325,
      "T": temperature
  },
  "broadcast_conditions": False,
  "output": "ZPF",
  "values":   values,
    "reference": reference,
    "comment": comment
}
with open(components[0]+"_"+components[1]+"_"+reference+"_"+"ZPF.json", "w") as outfile:
        json.dump(dictionary1, outfile,indent=1)
```
