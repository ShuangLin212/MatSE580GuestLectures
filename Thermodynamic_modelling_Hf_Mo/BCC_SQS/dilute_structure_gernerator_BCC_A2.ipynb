```python
import itertools
#Endmember structure generator
phase = 'BCC_A2'
title = 'sqs structures of '+phase+' generated by Shuang Lin'
binary_system = ['Hf','Mo']
original_file_name = 'bccsqs16_2575.txt'
line_0 = 'bccsqs16_25_75'
line_5 ='AA BB' # starting endmember combination in line 5 of oringinal poscar
line_6 = ' 12   4' # line 6
```

```python
counter =0
#subsititute first sublattice 
a = []
listOLists1 = [[binary_system[0]+' '+binary_system[1]]]
for i in itertools.product(*listOLists1):
    a.append(i)
listOLists2 = [[binary_system[1]+' '+binary_system[0]]]
for i in itertools.product(*listOLists2):
    a.append(i)
print(a)
for i in a:
    position_2a = i[0]
    a_file = open(original_file_name, "r")
    summary = open("summary.txt", "a")  # append mode
    summary.write(str(i))
    lines = a_file.readlines()
    lines[0]=lines[0].replace(line_0,title)
    lines[5]=lines[5].replace(line_5,position_2a)
    #print(lines[5])
    lines[6]=lines[6].replace(line_6,'12 4')
    for i in range(8,20):
        lines[i]=lines[i].replace(lines[i][-3]+lines[i][-2],position_2a[0]+position_2a[1])
    for i in range(20,24):
        lines[i]=lines[i].replace(lines[i][-3]+lines[i][-2],position_2a[3]+position_2a[4])
    counter = counter + 1
    #print(counter)
    summary.write(str(counter)+'\n')
    summary.close()
    a_file = open(str(counter)+"_"+phase+"_sqs_structure.POSCAR",'w',encoding='utf-8')
    a_file.writelines(lines)
    a_file.close()
```
