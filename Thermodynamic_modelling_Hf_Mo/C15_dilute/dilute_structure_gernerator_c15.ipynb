```python
import itertools
#Endmember structure generator
title = 'dilute structures of Laves_C15 generated by Shuang Lin'
phase = 'Laves_C15'
binary_system = ['Hf','Mo']
original_file_name = 'C15_dilute.POSCAR'
line_0 = 'Hf8 Mo16'
line_5 ='Hf Mo' # starting endmember combination in line 5 of oringinal poscar
line_6 = '16 32' # line 6
```

```python
counter =0
#subsititute first sublattice 
a = []
listOLists1 = [[binary_system[0]+' '+binary_system[1]],binary_system]
for i in itertools.product(*listOLists1):
    a.append(i)
listOLists2 = [[binary_system[1]+' '+binary_system[0]],binary_system]
for i in itertools.product(*listOLists2):
    a.append(i)
print(a)
for i in a:
    position_8a = i[0]
    position_16d = i[1]
    a_file = open(original_file_name, "r")
    summary = open("summary.txt", "a")  # append mode
    summary.write(str(i))
    lines = a_file.readlines()
    lines[0]=lines[0].replace(line_0,title)
    lines[5]=lines[5].replace(line_5,position_8a+' '+position_16d)
    #print(lines[5])
    lines[6]=lines[6].replace(line_6,'1 15 32')
    for i in range(8,9):
        lines[i]=lines[i].replace(lines[i][-7]+lines[i][-6],position_8a[0]+position_8a[1])
    for i in range(9,24):
        lines[i]=lines[i].replace(lines[i][-7]+lines[i][-6],position_8a[3]+position_8a[4])
    for i in range(24,56):
        lines[i]=lines[i].replace(lines[i][-7]+lines[i][-6],position_16d)
    counter = counter + 1
    #print(counter)
    summary.write(str(counter)+'\n')
    summary.close()
    a_file = open(str(counter)+"_"+phase+"_dilute_structure.POSCAR",'w',encoding='utf-8')
    a_file.writelines(lines)
    a_file.close()
```

```python
#subsititute the second sublattice 
b = []
listOLists3 = [binary_system,[binary_system[0]+' '+binary_system[1]]]
for i in itertools.product(*listOLists3):
    b.append(i)
listOLists4 = [binary_system,[binary_system[1]+' '+binary_system[0]]]
for i in itertools.product(*listOLists4):
    b.append(i)
print(b)
for i in b:
    position_8a = i[0]
    position_16d = i[1]
    a_file = open(original_file_name, "r")
    summary = open("summary.txt", "a")  # append mode
    summary.write(str(i))
    lines = a_file.readlines()
    lines[0]=lines[0].replace(line_0,title)
    lines[5]=lines[5].replace(line_5,position_8a+' '+position_16d)
    #print(lines[5])
    lines[6]=lines[6].replace(line_6,'16 1 31')
    for i in range(8,24):
        lines[i]=lines[i].replace(lines[i][-7]+lines[i][-6],position_8a)
    for i in range(24,25):
        lines[i]=lines[i].replace(lines[i][-7]+lines[i][-6],position_16d[0]+position_16d[1])
    for i in range(25,56):
        lines[i]=lines[i].replace(lines[i][-7]+lines[i][-6],position_16d[3]+position_16d[4])
    counter = counter + 1
    #print(counter)
    summary.write(str(counter)+'\n')
    summary.close()
    a_file = open(str(counter)+"_"+phase+"_dilute_structure.POSCAR",'w',encoding='utf-8')
    a_file.writelines(lines)
    a_file.close()
```
